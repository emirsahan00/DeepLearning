```python
import cv2  #gerekli ktphaneleri tanmlyoruz
import pickle
import numpy as np

def preprocess(img):  #modelle ayn formatta olmal
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  #resmi gri tona eviriyoruz
    img = cv2.equalizeHist(img) #histogram eitliyoruz 
    img = img/255  #ve normalize ediyoruz
    return img 

cap = cv2.VideoCapture(0)
cap.set(3,480)  #genilik ve ykseklik belirliyoruz
cap.set(4,480)

pickle_in = open('model_trained.p','rb')  #modeli ie aktaryoruz
model = pickle.load(pickle_in)

while True:  #sonsuz dng oluturuyoruz
    success ,frame =cap.read() #frameleri okuyoruz

    img = np.asarray(frame)  #numpy dizisi yapyoru
    img = cv2.resize(img,(32,32)) #32x32 olmak zere yeniden boyutlandryoruz
    img = preprocess(img) #preprocess fonksiyonu imgi yolluyoruz

    img = img.reshape(1,32,32,1) # 1 resim var 32 32 boyutlarnda ve tek kanall

    #predict -> tahmin 
    #classIndex = int(model.predict_classes(img))  -->eski srm
    classIndex = np.argmax(model.predict(img), axis=-1) # en yksek deerin indexini dndryoruz
    predictions = model.predict(img) 
    probVal = np.amax(predictions) #tahmin yzdesini probVal deikenine atyoruz
    print(classIndex,probVal)

    if probVal > 0.7:  #deer ihtimali %70'in stndeyse ekrana yazdr
        cv2.putText(frame,str(classIndex)+'   '+str(probVal),(50,50),cv2.FONT_HERSHEY_DUPLEX,1,(0,255,0),1)
    cv2.imshow('Rakam Siniflandirma',frame)
    if cv2.waitKey(1) == ord('q'): #eer ki 'q' tuuna baslrsa dngy kr
        break
cap.release()  #grnty serbest brak
cv2.destroyAllWindows() #tm pencereleri kapat
```

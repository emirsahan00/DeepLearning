```python
import cv2  #gerekli kütüphaneleri tanımlıyoruz
import pickle
import numpy as np

def preprocess(img):  #modelle aynı formatta olmalı
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  #resmi gri tona çeviriyoruz
    img = cv2.equalizeHist(img) #histogram eşitliyoruz 
    img = img/255  #ve normalize ediyoruz
    return img 

cap = cv2.VideoCapture(0)
cap.set(3,480)  #genişlik ve yükseklik belirliyoruz
cap.set(4,480)

pickle_in = open('model_trained.p','rb')  #modeli içe aktarıyoruz
model = pickle.load(pickle_in)

while True:  #sonsuz döngü oluşturuyoruz
    success ,frame =cap.read() #frameleri okuyoruz

    img = np.asarray(frame)  #numpy dizisi yapıyoru
    img = cv2.resize(img,(32,32)) #32x32 olmak üzere yeniden boyutlandırıyoruz
    img = preprocess(img) #preprocess fonksiyonu imgi yolluyoruz

    img = img.reshape(1,32,32,1) # 1 resim var 32 32 boyutlarında ve tek kanallı

    #predict -> tahmin 
    #classIndex = int(model.predict_classes(img))  -->eski sürüm
    classIndex = np.argmax(model.predict(img), axis=-1) # en yüksek değerin indexini döndürüyoruz
    predictions = model.predict(img) 
    probVal = np.amax(predictions) #tahmin yüzdesini probVal değişkenine atıyoruz
    print(classIndex,probVal)

    if probVal > 0.7:  #değer ihtimali %70'in üstündeyse ekrana yazdır
        cv2.putText(frame,str(classIndex)+'   '+str(probVal),(50,50),cv2.FONT_HERSHEY_DUPLEX,1,(0,255,0),1)
    cv2.imshow('Rakam Siniflandirma',frame)
    if cv2.waitKey(1) == ord('q'): #eğer ki 'q' tuşuna basılırsa döngüyü kır
        break
cap.release()  #görüntüyü serbest bırak
cv2.destroyAllWindows() #tüm pencereleri kapat
```

```python
import cv2
import matplotlib.pyplot as plt
import os 
from os.path import isfile,join

pathIn = r"img1"
pathOut = "D:/examples_forDeepLearning/ObjectTracking/KesifselVeriAnalizi/deneme.mp4"

files = [f for f in os.listdir(pathIn) if isfile(join(pathIn,f))] #resim dosyasının (img1) içinde gezinip files'a atıyoruz

# img = cv2.imread(pathIn + "//" + files[44])
# img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# plt.imshow(img)


#görüntüleri videoya dönüştürme kısmı 
fps = 25
size = (1920,1080)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*"MP4V"),fps,size,True)

for i in files:
    print(i)

    fileName = pathIn + "//" + i
    img = cv2.imread(fileName)
    out.write(img)

out.release()
```

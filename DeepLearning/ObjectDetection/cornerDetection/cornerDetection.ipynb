```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('D:/examples_forDeepLearning/ObjectDeteciton/cornerDetection/src/sudoku.png',0)
img = np.float32(img)
plt.figure(),plt.imshow(img,cmap="gray"),plt.axis("off")

dst = cv2.cornerHarris(img,blockSize=2,ksize=3,k=0.04) #köşeler bu işlemle bulunuyor 2 komşu bloğa bak 3 kutucuklu gezinti yap 
plt.figure(),plt.imshow(dst,cmap="gray"),plt.axis("off")

dst = cv2.dilate(dst,None)
img[dst>0.2*dst.max()] = 1

plt.figure(),plt.imshow(dst,cmap="gray"),plt.axis("off")

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('D:/examples_forDeepLearning/ObjectDeteciton/cornerDetection/src/sudoku.png',0)
img = np.float32(img)
corners = cv2.goodFeaturesToTrack(img,120,0.01,10) #src,100 köşe bul ,quailtyLevel , minimum mesafe
corners = np.int64(corners)

for i in corners:
    x,y = i.ravel()
    cv2.circle(img,(x,y),3,(125,125,125),cv2.FILLED) #  'cv2.FILLED' içi dolu bir çember

plt.imshow(img),plt.axis("off")


```

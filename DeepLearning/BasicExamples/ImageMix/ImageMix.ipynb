```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('D:/examples_forDeepLearning/ImageMix/src/manzara1.jpg')  #resimeri diskten ykledik
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)  #matplotlib ktphanesi kullandmz iin bgr formatndan rgb formatna evirdik 
img2  = cv2.imread('D:/examples_forDeepLearning/ImageMix/src/manzara2.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

img1 = cv2.resize(img1,(500,300)) #birletireceimiz resimlerin boyutlar ayn olmal 
img2 = cv2.resize(img2,(500,300))

blended = cv2.addWeighted(src1 = img1,alpha=0.5,src2= img2,beta= 0.5,gamma=0)  #hangi oranda birleeceklerini belirledik

plt.figure()  #matplotlib ktphanesi ile imshowladk
plt.imshow(img1)

plt.figure()
plt.imshow(img2)

plt.figure()
plt.imshow(blended)
```

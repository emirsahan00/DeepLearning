```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('D:/examples_forDeepLearning/ImageMix/src/manzara1.jpg')  #resimeri diskten yükledik
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)  #matplotlib kütüphanesi kullandığımız için bgr formatından rgb formatına çevirdik 
img2  = cv2.imread('D:/examples_forDeepLearning/ImageMix/src/manzara2.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

img1 = cv2.resize(img1,(500,300)) #birleştireceğimiz resimlerin boyutları aynı olmalı 
img2 = cv2.resize(img2,(500,300))

blended = cv2.addWeighted(src1 = img1,alpha=0.5,src2= img2,beta= 0.5,gamma=0)  #hangi oranda birleşeceklerini belirledik

plt.figure()  #matplotlib kütüphanesi ile imshowladık
plt.imshow(img1)

plt.figure()
plt.imshow(img2)

plt.figure()
plt.imshow(blended)
```
